### 作業2：
實際練習 GPIOZero 控制 LED，確定單獨的 GPIO 控制 LED 亮跟暗交替閃爍完成，並且 PWM 控制 LED 明亮的完成後，嘗試依序改變led.value的值，分別設定 0.1, 0.3, 0.5, 0.7 觀察差異。

```python
from gpiozero import LED
from time import sleep
led = LED(19)
led.on()
sleep(5)
led.off()
sleep(5)
```

```python
from gpiozero import PWMLED
led = PWMLED(19)

while True:
    led.value = 0.1
    sleep(1)
    led.value = 0.3
    sleep(1)
    led.value = 0.5
    sleep(1)
    led.value = 0.7
    sleep(1)
    
```

### 作業3：
實際練習 GPIOZero 透過 Button 控制 LED，在按鈕的過程中，觀察實際按下按鈕的次數，LED 點亮的次數，是否一致。練習修改程式，讓按鈕按下是全亮，按鈕放開後是 30% 的亮度。

```python
from gpiozero import Button
from gpiozero import PWMLED

btn = Button(26)
led = PWMLED(19)
sum = 0

while True:
    if btn.is_pressed:
        led.value = 1
        sum = sum +1
        print(f"LED Light on times : {sum}")
    
    else
        led.value =0.3
```
